```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tokenize import sent_tokenize
from nltk.tokenize import TweetTokenizer
from nltk.probability import FreqDist
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.sentiment.vader import SentimentIntensityAnalyzer

#this is sample data
from nltk.corpus import names  

from string import punctuation
```

```python
princess=[]

with open('12dancingprincesses.txt', 'r') as f:
    for line in f:
        cline = line.strip() #get rid of newline character
        if cline == '': pass  #this will skip over lines that only had a newline and are now blank
        else:
            tknls = word_tokenize(cline)

        for token in tknls:
            princess.append(token)
            
```

```python
new_punct = punctuation + "’" + "‘"
```

```python
#remove the puntuation tokens from the list
for token in princess:
    if token in new_punct:
        princess.remove(token)
```

```python
eng_stopwords = stopwords.words('english')
eng_stopwords
```

```python
rm_count = 0
new_princess = []  #list to hold new words

for token in princess:
    if token not in eng_stopwords:
        new_princess.append(token)
    else: rm_count += 1
```
